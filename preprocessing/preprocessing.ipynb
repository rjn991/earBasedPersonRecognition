```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import cv2
```

```python
img= cv2.imread('04.png')[:,:,::-1]
```

```python
type (img)
```

```python
print(img.shape)
```

## Original

```python
plt.imshow(img)
plt.show()
```

## Gaussian Blur

```python
output_gaus = cv2.GaussianBlur(img,(7,7),0)
img_plot=plt.imshow(output_gaus)
plt.show()
```

## Median Blur

```python
med_blur = cv2.medianBlur(img,5)
img_plot=plt.imshow(med_blur)
plt.show()
```

## Denoising

```python
dst = cv2.fastNlMeansDenoisingColored(img,None,9,9,7,21)
img_plot=plt.imshow(dst)
plt.show()
```

# Sharpening

```python
sharp = cv2.addWeighted(img, 1.5, output_gaus, -0.5,0);
img_plot=plt.imshow(sharp)
plt.show()
```

```python
sharp2 = cv2.addWeighted(img, 1.5, med_blur, -0.5,0);
img_plot=plt.imshow(sharp2)
plt.show()
```
